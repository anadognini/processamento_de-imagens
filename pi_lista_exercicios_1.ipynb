```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Gustavo Westphal Antunes, Ana Dognini
```

```python
def exemplo_01():
    img = cv2.imread('/content/sample_data/LenaC.jpg')

    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Imagem Colorida')
    plt.show()

    cv2.imwrite('/content/sample_data/LenaC1.jpg', img)

exemplo_01()
```

```python
def exemplo_02():
    img = cv2.imread('/content/sample_data/LenaC.jpg')

    imagem_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Imagem Original')

    plt.subplot(122)
    plt.imshow(imagem_cinza, cmap='gray')
    plt.title('Imagem em Níveis de Cinza')

    plt.tight_layout()
    plt.show()

    cv2.imwrite('/content/sample_data/lena_cinza.jpg', imagem_cinza)


exemplo_02()
```

```python
def exemplo_03():
    img = cv2.imread('/content/sample_data/LenaC.jpg')

    canais = cv2.split(img)

    img_r = cv2.merge([canais[2], np.zeros_like(canais[1]), np.zeros_like(canais[0])])
    img_g = cv2.merge([np.zeros_like(canais[2]), canais[1], np.zeros_like(canais[0])])
    img_b = cv2.merge([np.zeros_like(canais[2]), np.zeros_like(canais[1]), canais[0]])

    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.imshow(cv2.cvtColor(img_r, cv2.COLOR_BGR2RGB))
    plt.title('Canal R')

    plt.subplot(1,3,2)
    plt.imshow(cv2.cvtColor(img_g, cv2.COLOR_BGR2RGB))
    plt.title('Canal G')

    plt.subplot(1,3,3)
    plt.imshow(cv2.cvtColor(img_b, cv2.COLOR_BGR2RGB))
    plt.title('Canal B')

    plt.tight_layout()
    plt.show()

    cv2.imwrite('/content/sample_data/lena_canal_R.jpg', img_r)
    cv2.imwrite('/content/sample_data/lena_canal_G.jpg', img_g)
    cv2.imwrite('/content/sample_data/lena_canal_B.jpg', img_b)

exemplo_03()
```

```python
def exemplo_04():
    img = cv2.imread('/content/sample_data/LenaC.jpg')
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img_media = cv2.blur(img_gray, (5,5))
    img_mediana = cv2.medianBlur(img_gray, 5)

    plt.figure(figsize=(15,5))
    plt.subplot(131)
    plt.imshow(img_gray, cmap='gray')
    plt.title('Imagem Original')

    plt.subplot(132)
    plt.imshow(img_media, cmap='gray')
    plt.title('Filtro Média')

    plt.subplot(133)
    plt.imshow(img_mediana, cmap='gray')
    plt.title('Filtro Mediana')

    plt.tight_layout()
    plt.show()

    cv2.imwrite('/content/sample_data/lena_media.jpg', img_media)
    cv2.imwrite('/content/sample_data/lena_mediana.jpg', img_mediana)

exemplo_04()
```

```python
def exemplo_05():
    img = cv2.imread('/content/sample_data/LenaC.jpg')
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, img_binary = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(img_gray, cmap='gray')
    plt.title('Imagem Original')

    plt.subplot(122)
    plt.imshow(img_binary, cmap='gray')
    plt.title('Imagem Binarizada')

    plt.tight_layout()
    plt.show()

    cv2.imwrite('/content/sample_data/lena_binarizada.jpg', img_binary)

exemplo_05()
```

```python
def exemplo_06():
    img = cv2.imread('/content/sample_data/LenaC.jpg')
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img_equalizada = cv2.equalizeHist(img_gray)
    plt.figure(figsize=(15,10))

    plt.subplot(221)
    plt.imshow(img_gray, cmap='gray')
    plt.title('Imagem Original')

    plt.subplot(222)
    plt.hist(img_gray.ravel(), 256, [0,256])
    plt.title('Histograma Original')

    plt.subplot(223)
    plt.imshow(img_equalizada, cmap='gray')
    plt.title('Imagem Equalizada')

    plt.subplot(224)
    plt.hist(img_equalizada.ravel(), 256, [0,256])
    plt.title('Histograma Equalizado')

    plt.tight_layout()
    plt.show()

exemplo_06()
```
