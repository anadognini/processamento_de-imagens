## Analysis of Image Processing Using Morphological Erosion and Dilation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread('/content/mãos.jpg', cv2.IMREAD_GRAYSCALE) # carrega a imagem
```

```python
_, imagem_bin = cv2.threshold(imagem, 127, 255, cv2.THRESH_BINARY) # binariza a imagem
elemento_estruturante = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
```

```python
erosao = cv2.erode(imagem_bin, elemento_estruturante, iterations=1) # aplica erosao

dilatacao = cv2.dilate(imagem_bin, elemento_estruturante, iterations=1) # aplica dilataçao

abertura = cv2.morphologyEx(imagem_bin, cv2.MORPH_OPEN, elemento_estruturante) # aplica abertura

fechamento = cv2.morphologyEx(imagem_bin, cv2.MORPH_CLOSE, elemento_estruturante) # aplica fechamento

gradiente = cv2.morphologyEx(imagem_bin, cv2.MORPH_GRADIENT, elemento_estruturante) # aplica gradiente morfologico
```

```python
# imprimir imagens

titulos = ['Original', 'Abertura', 'Fechamento', 'Erosão', 'Gradiente' ]
imagens = [imagem_bin, abertura, fechamento, erosao, gradiente]
plt.figure(figsize=(20, 5))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(imagens[i], cmap='gray')
    plt.title(titulos[i])
    plt.axis('off')

plt.show()
```
