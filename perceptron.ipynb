```python
neymar = [1, 0, 0, 0]
messi = [1, 0, 1, 0]
barrichello = [1, 1, 0, 1]
massa = [1, 1, 1, 1]
```

```python
def funcao_ativacao(soma):
  if soma >= 1:
    return 1
  else:
    return 0
```

```python
arr = [neymar, messi, barrichello, massa]
peso = [0, 0, 0]
processados = []
```

```python
while True:
  for player in arr:


    soma = 0

    for i in range(3):
      soma = player[i] * peso[i]

    if (funcao_ativacao(soma) == player[3]):
      print("Pertence à classe %i" % player[3])
    else:
      processados.append(player)

      if soma == 0:
        for j in range(len(peso)):
          peso[i] += player[j]
      else:
        for j in range(len(peso)):
          peso[i] -= player[j]

      print("Não pertence à classe %i" % player[3])

  if len(processados) > 0:
      processados = []
      continue
  else:
      break
```
